## Using numpy

````python
# 1. Declare a 3x3 matrix
matrix_a_np = np.array([[1, 4, 3],
                        [4, 5, 6],
                        [7, 8, 9]])
print("\nMatrix A (NumPy):\n", matrix_a_np)
````

````python
# 2. Add two 3x3 matrices
matrix_b_np = np.array([[9, 2, 7],
                        [6, 5, 4],
                        [3, 2, 1]])
matrix_sum_np = matrix_a_np + matrix_b_np

print("\nMatrix B (NumPy):\n", matrix_b_np)
print("\nSum of Matrix A and Matrix B (NumPy):\n", matrix_sum_np)
````

````python
# 3. Subtract two 3x3 matrices
matrix_diff_np = matrix_a_np - matrix_b_np

print("\nDifference of Matrix A and Matrix B (NumPy):\n", matrix_diff_np)

````

````python
# 4. Multiply a matrix by a number (scalar multiplication)
scalar = 2
matrix_scalar_mult_np = scalar * matrix_a_np
print(f"\nMatrix A multiplied by {scalar} (NumPy):\n", matrix_scalar_mult_np)
````

````python
# 5. Multiply a matrix by another matrix (matrix multiplication)
matrix_mult_np = np.dot(matrix_a_np, matrix_b_np)
# Alternatively, you can use the @ operator in Python 3.5+
# matrix_mult_np = matrix_a_np @ matrix_b_np
print("\nProduct of Matrix A and Matrix B (NumPy):\n", matrix_mult_np)
````

````python
# 6. Transpose of a matrix
matrix_transpose_np = matrix_a_np.T
print("\nTranspose of Matrix A (NumPy):\n", matrix_transpose_np)
````

````python
# 7. Inverse a matrix

try:
  matrix_inverse_np = np.linalg.inv(matrix_a_np)
  print("\nInverse of Matrix A (NumPy):\n", matrix_inverse_np)
except np.linalg.LinAlgError:
  print("\nMatrix A is singular and does not have an inverse.")
````

````python
# Singular Value Decomposition (SVD)

# Matrix U, (array) S(igma) and V
U, S, V = np.linalg.svd(matrix_a_np)

print("\nU (left singular vectors):\n", U)
print("\nS (singular values):\n", S)
print("\nV (right singular vectors):\n", V)

# U and V are orthogonal matrices

# prod1 = np.dot(U, U.transpose())
# print(prod1)
# prod2 = np.dot(U.transpose(), U)
# print(prod2)

# prod1 = np.dot(V, V.transpose())
# print(prod1)
# prod2 = np.dot(V.transpose(), V)
# print(prod2)

````

## What is Covariance


Covariance is a measure of how much two random variables change together.

It indicates the direction of the linear relationship between the variables.

Mathematically, the covariance between two random variables $X$ and $Y$ is often denoted as $cov(X, Y)$ or $\sigma_{XY}$.

One way to define covariance is through its relationship with the Pearson correlation coefficient ($\rho_{X,Y}$), as shown in the image you provided:

$$ \rho_{X,Y} = \frac{cov(X, Y)}{\sigma_X \sigma_Y} $$



Where:
* $\rho_{X,Y}$ is the Pearson correlation coefficient between $X$ and $Y$.
* $cov(X, Y)$ is the covariance between $X$ and $Y$.
* $\sigma_X$ is the standard deviation of $X$.
* $\sigma_Y$ is the standard deviation of $Y$.


By definition, the sample covariance for $n$ data points $(x_i, y_i)$ is calculated as:

$$ cov(X, Y) = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{n-1} $$

Where:
* $x_i$ and $y_i$ are the individual data points for variables $X$ and $Y$.
* $\bar{x}$ is the mean (average) of variable $X$.
* $\bar{y}$ is the mean (average) of variable $Y$.
* $n$ is the number of data points.

We devide by $n-1$, not $n$ on the LHS of the definition. The number ONE (as in n - 1) is called degree of freedom.

**Interpretation:**

* **Positive Covariance ($cov(X, Y) > 0$):** Indicates that as one variable increases, the other tends to increase as well.
* **Negative Covariance ($cov(X, Y) < 0$):** Indicates that as one variable increases, the other tends to decrease.
* **Zero Covariance ($cov(X, Y) = 0$):** Indicates that there is no linear relationship between the two variables. They vary independently of each other.

When $abs(cov(X,Y))$ is greater than 0.8, we say $X$ and $Y$ are strongly related.

**Note:** The correlation coefficient (value in the range $[-1, 1]$), which is a normalized version of covariance, provides a more standardized measure of the linear relationship. The magnitude of covariance is not easily interpretable as it depends on the scales of the variables.

## What Correlation Coefficients Signify

````python



# Set random seed for reproducibility
np.random.seed(0)

# Number of data points
n = 100

# Correlation coefficients
rhos = [1, 0.8, 0.4, 0, -0.4, -0.8, -1]

# Create subplots
fig, axes = plt.subplots(1, 7, figsize=(21, 7)) # Increased the height to 7

for i, rho in enumerate(rhos):
    # Generate correlated variables
    X = np.random.normal(0, 1, n)
    Y = rho * X + np.random.normal(0, np.sqrt(1 - rho**2), n)

    # Calculate regression line
    m, b = np.polyfit(X, Y, 1)

    # Plot data and regression line
    axes[i].scatter(X, Y, label=f'Correlation = {rho}')
    axes[i].plot(X, m * X + b, color='red', label=f'Regression line (y = {m:.2f}x + {b:.2f})')
    axes[i].set_xlabel('X')
    axes[i].set_ylabel('Y')
    axes[i].set_title(f'Corr Coefficient = {rho}')
    # axes[i].legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Moved legend outside the graph
    axes[i].set_xlim([-4, 4])
    axes[i].set_ylim([-4, 4])
    axes[i].set_aspect('equal', adjustable='box') # Ensure equal scaling for x and y axes


plt.tight_layout(rect=[0, 0, 0.95, 1]) # Adjust layout to prevent overlap with legends
plt.show()
````

# SVD (Singular Value Decomposition)


The Singular Value Decomposition (SVD) is a matrix factorization technique that decomposes any real or complex matrix $\mathbf{A}$ into the product of three other matrices: an orthogonal matrix $\mathbf{U}$, a diagonal matrix $\mathbf{\Sigma}$, and the transpose of an orthogonal matrix $\mathbf{V}$.

For an $m \times n$ matrix $\mathbf{A}$, the SVD is given by:

$$ \mathbf{A} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T $$

Where:

* $\mathbf{U}$ is an $m \times m$ orthogonal matrix whose columns are the **left singular vectors** of $\mathbf{A}$. These are the eigenvectors of $\mathbf{A} \mathbf{A}^T$.
* $\mathbf{\Sigma}$ is an $m \times n$ diagonal matrix with non-negative real numbers on the diagonal, known as the **singular values** of $\mathbf{A}$, sorted in descending order. $\mathbf{\Sigma}$ has the form:

$$ \mathbf{\Sigma} =
\begin{pmatrix}
    \sigma_1 & 0 & \cdots & 0 \\
    0 & \sigma_2 & \cdots & 0 \\
    \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & \cdots & \sigma_r \\
    0 & 0 & \cdots & 0 \\
    \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & \cdots & 0
\end{pmatrix}
$$

where $\sigma_1 \ge \sigma_2 \ge \cdots \ge \sigma_r > 0$, and $r$ is the rank of $\mathbf{A}$.

* $\mathbf{V}$ is an $n \times n$ orthogonal matrix whose columns are the **right singular vectors** of $\mathbf{A}$. These are the eigenvectors of $\mathbf{A}^T \mathbf{A}$.

**Key Properties of SVD:**

* $\mathbf{U}^T \mathbf{U} = \mathbf{I}_m$ (where $\mathbf{I}_m$ is the $m \times m$ identity matrix)
* $\mathbf{V}^T \mathbf{V} = \mathbf{I}_n$ (where $\mathbf{I}_n$ is the $n \times n$ identity matrix)
* The singular values $\sigma_i$ are the square roots of the non-zero eigenvalues of both $\mathbf{A} \mathbf{A}^T$ and $\mathbf{A}^T \mathbf{A}$.


## SVD Contributes to Dimensionality Reduction in PCA


In PCA, we select the top $k$ principal components (eigenvectors with the largest eigenvalues) to reduce the dimensionality of the data while *retaining the most variance*. Similarly, using SVD for dimensionality reduction involves selecting the top $k$ singular values and their corresponding left and right singular vectors. We can then reconstruct an approximation of the original matrix $\mathbf{A}$ using only these top $k$ components:

$$ \mathbf{A}_k \approx \mathbf{U}_k \mathbf{\Sigma}_k \mathbf{V}_k^T $$
where $\mathbf{U}_k$ consists of the first $k$ columns of $\mathbf{U}$, $\mathbf{\Sigma}_k$ is the top-left $k \times k$ diagonal matrix of singular values, and $\mathbf{V}_k$ consists of the first $k$ columns of $\mathbf{V}$.

With scikit-learn, we can specify the number of dimensions to reduce on initilisation of PCA object.

## Application of SVD in PCA for Movie Recommendations

**Scenario:**

Consider a dataset where we have $n$ users and $d$ movies. Let matrix $\mathbf{R}$ with elements $r_{ij}$ represent the rating given by user $i$ to movie $j$. We hypothesize that a user's movie preferences are actually determined by a smaller number of $k$ underlying factors, such as movie genres (e.g., action, comedy, drama), preferred actors, or preferred directors. An individual user's rating behavior is then a weighted combination of these underlying factors.

**Applying SVD:**

We can apply SVD to the user-movie rating matrix $\mathbf{R}$ (of size $n \times d$):

$$ \mathbf{R} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T $$

Where:

* $\mathbf{U}$ is an $n \times n$ orthogonal matrix. In the context of PCA for recommendations, the first $k$ columns of $\mathbf{U}$ (an $n \times k$ matrix) can be seen as representing the projection of each user onto the $k$ principal components (underlying factors). Each row in this $n \times k$ matrix represents a user, and the values in that row are the weights of that user's preference for each of the $k$ underlying factors.

* $\mathbf{\Sigma}$ is an $n \times d$ diagonal matrix with non-negative singular values $\sigma_1 \ge \sigma_2 \ge \dots \ge \sigma_{\min(n,d)} \ge 0$ on the diagonal. These singular values represent the importance or variance captured by each corresponding principal component (underlying factor).

* $\mathbf{V}^T$ is a $d \times d$ orthogonal matrix. The first $k$ columns of $\mathbf{V}$ (so $\mathbf{V}^T$ will have the first $k$ rows, forming a $k \times d$ matrix) can be seen as representing the association of each movie with the $k$ principal components. Each column in the original $d \times k$ matrix $\mathbf{V}$ (or each row in the $k \times d$ matrix $\mathbf{V}^T$) represents a principal component (underlying factor), and the values in that row (or column) indicate how much each movie is related to that particular factor.

**Rank-$k$ Approximation and PCA:**

Similar to the customer-product example, we can take the best rank-$k$ approximation of $\mathbf{R}$, denoted as $\mathbf{R}_k$, by using only the top $k$ singular values and their corresponding singular vectors:

$$ \mathbf{R}_k = \mathbf{U}_k \mathbf{\Sigma}_k \mathbf{V}_k^T $$

Where $\mathbf{U}_k$ is the $n \times k$ matrix of the first $k$ columns of $\mathbf{U}$, $\mathbf{\Sigma}_k$ is the $k \times k$ diagonal matrix of the top $k$ singular values, and $\mathbf{V}_k^T$ is the $k \times d$ matrix of the first $k$ rows of $\mathbf{V}^T$ (or equivalently, the transpose of the first $k$ columns of $\mathbf{V}$).

In this application of SVD to PCA for movie recommendations:

* The columns of $\mathbf{U}_k$ represent the principal components in the "user space". Each user is now represented by a $k$-dimensional vector of weights corresponding to these principal components.
* The columns of $\mathbf{V}_k$ (or rows of $\mathbf{V}_k^T$) represent the principal components in the "movie space". Each movie is also represented by a $k$-dimensional vector indicating its association with these principal components.
* The singular values in $\mathbf{\Sigma}_k$ indicate the importance of each of these $k$ principal components in explaining the variance in user ratings.

By reducing the dimensionality of both users and movies to $k$ factors, we can then perform various recommendation tasks, such as:

* **Predicting missing ratings:** We can estimate the rating of user $i$ for movie $j$ using the rank-$k$ approximation $\mathbf{R}_k$.
* **Finding similar users:** Users with similar weight vectors in $\mathbf{U}_k$ are likely to have similar movie preferences.
* **Finding similar movies:** Movies with similar association vectors in $\mathbf{V}_k$ are likely to be liked by the same users.

This application mirrors the concept in Section 1.5.1 where customer purchase behavior is linked to underlying factors, and SVD is used to uncover these factors and represent both customers and products in a lower-dimensional space. In our movie recommendation example, user ratings are linked to underlying movie preference factors, and SVD helps us represent users and movies in a reduced dimensional space of these factors.
