# Sigmoid Function


Logistic sigmoid function,  denoted by $\sigma(x)$ or $S(x)$:

$$ S(x) = \sigma(x) = \frac{1}{1 + e^{-x}}$$

Where:
* $x$ is the input value (a real number).
* $e$ is the base of the natural logarithm (approximately 2.71828).

$\sigma(x)$ takes values from 0 to 1. The shape of the looks like like the character "S".

Sigmoid function is widely used in logistic regression, Neural Networks as an activation function.

# Logistic regression

## Maths Background

Let's explain those points mathematically in a short and simple way for Logistic Regression:

1.  **Weighted Sum + Bias:**

For an input $\mathbf{x}$ with features $[x_1, x_2, \dots, x_n]$ and parameters $\mathbf{\theta}$ (weights $[\theta_1, \theta_2, \dots, \theta_n]$ and bias $\theta_0$), we calculate the linear score $z$:

$$z = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n$$
  
  In vector form: $z = \mathbf{\theta}^T \mathbf{x}$ (assuming $\mathbf{x}$ includes a bias term of 1).

2.  **Estimating Probability ($\hat{p}$):**
    We pass the linear score $z$ through the Sigmoid function to get the estimated probability $\hat{p}$ of the positive class (class 1):
    $$\hat{p} = \sigma(z) = \frac{1}{1 + e^{-z}}$$
    $\hat{p}$ is always between 0 and 1.

3.  **Classification Decision:**
    Based on the estimated probability $\hat{p}$, we classify the instance:

$$ \hat{y} = \begin{cases} 0 & \text{if } \hat{p} < 0.5 \\ 1 & \text{if } \hat{p} \ge 0.5 \end{cases}$$

The threshold 0.5 corresponds to $z=0$. If $z > 0$, $\hat{p} > 0.5$; if $z < 0$, $\hat{p} < 0.5$.

4.  **Training Goal (Finding $\mathbf{\theta}$):**
    The goal is to find the parameter vector $\mathbf{\theta}$ that makes $\hat{p}$ high for positive instances (true label $y=1$) and low for negative instances ($y=0$).

5.  **How to Find $\mathbf{\theta}$ (Gradient Descent):**
    We define a **Cost Function** (typically Cross-Entropy between $p$ and $\hat{p}$) that measures the error between the predicted probabilities $\hat{p}$ and the true labels $y$ for the entire dataset.
    Gradient Descent is an iterative optimization algorithm that starts with initial $\mathbf{\theta}$ and repeatedly updates $\mathbf{\theta}$ in the direction that decreases the cost function (the negative of the gradient). This process aims to find the $\mathbf{\theta}$ that minimizes the cost function, leading to the best fit for the data.

# Cost Function in Logistic Regression

## Loss function

$$y^{(i)} \log(\hat{p}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{p}^{(i)})$$

for a single training example.

## Cost function

$$J(\theta) = - \frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \log(\hat{p}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{p}^{(i)})]$$

is the cost function for entire training dataset. The goal of training is to find the $\theta$ that minimizes this cost.

## Simple Logistic Regression Demo

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# X: features (e.g., two features for simplicity)
# y: labels (0 or 1 for binary classification)
X = np.array([
    [1.0, 2.0],
    [1.5, 1.8],
    [5.0, 8.0],
    [8.0, 8.0],
    [1.0, 0.6],
    [9.0, 11.0],
    [2.0, 3.0],
    [6.0, 10.0]
])

# Binary target
y = np.array([0, 0, 1, 1, 0, 1, 0, 1])

# Create and train the Logistic Regression model
# Instantiate the model
model = LogisticRegression()
# TODO Try to set hyperparams in LogisticRegression()

# Train the model using the hardcoded data
model.fit(X, y)

print("Logistic Regression model trained with hardcoded data.")

# The model is now trained and ready for predictions if needed.
# You can inspect the learned coefficients and intercept:
print("\nLearned coefficients:", model.coef_)
print("Learned intercept:", model.intercept_)
```

## Build a binary classifier to detect the Iris virginica

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
# based only on the petal width feature
X = iris["data"][:, 3:]  # petal width
y = (iris["target"] == 2).astype(np.int64)  # 1 if Iris virginica, else 0
# y = (iris["target"] == 2).astype(np.int)  # 1 if Iris virginica, else 0
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver="lbfgs", random_state=42)
log_reg.fit(X, y)
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris["data"][:, 3:]  # petal width
y = (iris["target"] == 2).astype(int)  # 1 if Iris virginica, else 0

# Train a logistic regression classifier
log_reg = LogisticRegression()
log_reg.fit(X, y)

# A fancy plot
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

# Calculate the decision boundary
filtered_indices = y_proba[:, 1] >= 0.5
if any(filtered_indices):
    decision_boundary = X_new[filtered_indices][0][0]
else:
    decision_boundary = None

plt.figure(figsize=(8, 3))
plt.plot(X[y==0], y[y==0], "bs")  # blue squares
plt.plot(X[y==1], y[y==1], "g^")  # green triangles

if decision_boundary is not None:
    plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
    plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
    plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
    plt.text(decision_boundary+0.02, 0.15, "Decision boundary", fontsize=14, color="k", ha="center")

plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])
plt.title("Logistic Regression - Probability of Iris Virginica")
plt.show()
```

## Logistic Classifier with two features: petal width and length

```python
from sklearn.linear_model import LogisticRegression

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.int64)  # 1 if Iris virginica, else 0

# y = (iris["target"] == 2).astype(np.int)

log_reg = LogisticRegression(solver="lbfgs", C=10**10, random_state=42)
log_reg.fit(X, y)

x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]

y_proba = log_reg.predict_proba(X_new)

plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)

left_right = np.array([2.9, 7])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]

plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.text(3.5, 1.5, "Not Iris virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])
plt.title("logistic_regression_contour_plot")
plt.show()
```

## Soft regression contour plot

```python
# X = iris["data"][:, (2, 3)]  # petal length, petal width
# y = iris["target"]

# softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs", C=10, random_state=42)
# softmax_reg.fit(X, y)

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

# Remove the deprecated 'multi_class' parameter
softmax_reg = LogisticRegression(solver="lbfgs", C=10, random_state=42)
softmax_reg.fit(X, y)
```

```python
x0, x1 = np.meshgrid(
        np.linspace(0, 8, 500).reshape(-1, 1),
        np.linspace(0, 3.5, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]


y_proba = softmax_reg.predict_proba(X_new)
y_predict = softmax_reg.predict(X_new)

zz1 = y_proba[:, 1].reshape(x0.shape)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Iris virginica")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Iris versicolor")
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Iris setosa")

from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])

plt.contourf(x0, x1, zz, cmap=custom_cmap)
contour = plt.contour(x0, x1, zz1, cmap=plt.cm.brg)
plt.clabel(contour, inline=1, fontsize=12)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 7, 0, 3.5])
plt.title("softmax_regression_contour_plot")
plt.show()
```
