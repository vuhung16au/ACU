## Data Cleaning

### Fill these missing values

Ref: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

```python
housing_ds['total_bedrooms'].mean() # fill with mean value
```

```python
housing_fillna = housing_ds['total_bedrooms'].fillna((housing_ds['total_bedrooms'].mean()))
```

```python
housing_fillna.isnull().sum()
```

```python
housing_fillna[288:293] # filled with mean value: 537.87 (please check housing.describe())
```

## Handling Text and Categorical Attributes

```python
housing_cat = housing[['ocean_proximity']] # Each value represents a category

housing_cat[1635:1642]
```

### Solution: One-hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
```

```python
housing_cat_1hot = cat_encoder.fit_transform(housing_cat)
```

```python
housing_cat_1hot_array = housing_cat_1hot.toarray()
housing_cat_1hot_array
```

```python
cat_encoder.categories_
```

```python
housing_cat[1635:1642]
```

```python
housing_cat_1hot_array[1635:1642]
```
